```python
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
import numpy as np
import string



'''

matriks awal

'''
# point a,b,c,d
a = (0,0,1)
b = (0,1,1)
c = (-1,1,1)
d = (-1,0,1)

X = np.array([a,b,c,d])



```

```python
'''
untuk ngegambar titiknya
'''


def sebelum (matrix):
    fg = plt.figure()
    ax = plt.gca()
    xs = []
    ys = []
    for row in matrix:
        x, y, i = row 
        xs.append(x)
        ys.append(y)
        i = int(i)
        plt.scatter(x, y)
        plt.text(x + 0.15, y, f"{string.ascii_letters[i]}")

    xs.append(xs[0])
    ys.append(ys[0])

    plt.plot(xs, ys, color="red", linestyle='dotted')
    ax.set_xticks(np.arange(-2.5, 3, 0.5))
    ax.set_yticks(np.arange(-2.5, 3, 0.5))
    plt.grid()
    plt.title("matriks sebelum")
    plt.show()
    
    
def sesudah (matrix):
    fig = plt.figure()
    ax = plt.gca()
    xs = []
    ys = []
    for row in matrix:
        x, y, i = row
        xs.append(x)
        ys.append(y)
        i = int(i)
        plt.scatter(x, y)
        plt.text(x + 0.15, y, f"{string.ascii_letters[i]}")

    xs.append(xs[0])
    ys.append(ys[0])

    plt.plot(xs, ys, color="red", linestyle='dotted')
    ax.set_xticks(np.arange(-2.5, 3, 0.5))
    ax.set_yticks(np.arange(-2.5, 3, 0.5))
    plt.grid()
    plt.title("matriks setelah")
    plt.show()
```

### TRANSLASI (TRANSLATION)

```python
'''
translasi
'''


def B (mtx, p1, p2):
    mat = np.array([
        [1, 0, p1],
        [0, 1, p2],
        [0, 0, 1]])
        
    tr = mtx.transpose()
    hasil = (mat @ tr)
    hasil = hasil.transpose()
    return np.array(hasil)


    
```

```python
translasi = B(X,-1,-1)
sebelum(X)
sesudah(translasi)
```

### SKALA (SCALE)


```python
def C (mtx,c1,c2):
    scale = np.array([
        [c1, 0, 0],
        [0, c2, 0],
        [0, 0, 1]
    ])
    
    sc = mtx.transpose()
    hasil = (scale @ sc)
    hasil = hasil.transpose()   
    return np.array(hasil)
```

```python
scaling = C(X,2,2)
sebelum(X)
sesudah(scaling)
```

```python
def E (mtx, e1, e2):
    shear = np.array([
        [1, e1, 0],
        [e2, 1, 0],
        [0, 0, 1]])
        
    sh = mtx.transpose()
    hasil = (shear @ sh)
    hasil = hasil.transpose()
    return np.array(hasil)
```

### PROYEKSI

```python
def F(matrix, f1,f2):
    m_projection = np.array([
            [1, 0, 0],
            [0, 1, 0],
            [f1, f2, 1]])

    m_t = matrix.transpose()
    hasil = m_projection @ m_t
    hasil = hasil / hasil[2]
    hasil = hasil.transpose()    

    return np.array(hasil)
```
